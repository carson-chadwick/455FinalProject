```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

df = pd.read_csv('shared_articles.csv')
df.head()
# Import google.translate for the portugues??
```

```python
df.dropna(inplace=True)
```

```python
# Step 1: Import/build your stop word dictionary

# Step 2: Create a TfidfVectorized and remove stop words
tfidf = TfidfVectorizer(stop_words='english')
# Step 3: Fit and transform the data to a tfidf matrix
tfidf_matrix = tfidf.fit_transform(df['text'])

print(tfidf_matrix)
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

df_results = pd.DataFrame(cosine_sim, index=df['contentId'], columns=df['contentId'])
df_results.sort_values(by=[6031953227014493100], ascending=False)
```

```python
def recommend_content(content_id, top_k=5):
    try:
        if content_id not in df_results.index:
            print(f"Content ID {content_id} not found in similarity matrix.")
            return []

        sim_scores = df_results.loc[content_id]

        # If it's the wrong shape, print and skip
        if not isinstance(sim_scores, pd.Series):
            print(f"sim_scores for {content_id} is not a Series. Type: {type(sim_scores)}")
            return []

        # Sort and drop self
        sorted_scores = sim_scores.sort_values(ascending=False)
        if content_id in sorted_scores.index:
            sorted_scores = sorted_scores.drop(content_id)

        top_ids = sorted_scores.head(top_k).index.tolist()
        return top_ids

    except Exception as e:
        print(f"Error in recommend_content({content_id}): {type(e).__name__} - {e}")
        return []

```

```python
print("df_results index sample:", df_results.index[:5])
print("First contentId in loop:", df['contentId'].iloc[0])
```

```python
duplicates = df['contentId'].duplicated().sum()
print(f"Number of duplicated contentIds in df: {duplicates}")
```

```python
results = []

for content_id in df['contentId']:
    try:
        rec_ids = recommend_content(content_id, top_k=5)
        titles = [df.loc[df['contentId'] == rid, "title"].values[0] if rid in df['contentId'].values else "Unknown" for rid in rec_ids]
        row = {"contentId": content_id}
        for i, title in enumerate(titles):
            row[f"recommendationTitle{i+1}"] = title
        results.append(row)
    except Exception as e:
        print(f"Skipping contentId {content_id} due to error: {e}")

# Save to CSV
pd.DataFrame(results).to_csv("ContentFiltering.csv", index=False)
```

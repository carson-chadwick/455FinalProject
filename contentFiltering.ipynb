```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

df = pd.read_csv('shared_articles.csv')
df.head()
# Import google.translate for the portugues??
```

```python
df.dropna(inplace=True)
```

```python
# Step 1: Import/build your stop word dictionary

# Step 2: Create a TfidfVectorized and remove stop words
tfidf = TfidfVectorizer(stop_words='english')
# Step 3: Fit and transform the data to a tfidf matrix
tfidf_matrix = tfidf.fit_transform(df['text'])

print(tfidf_matrix)
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

df_results = pd.DataFrame(cosine_sim, index=df['contentId'], columns=df['contentId'])
df_results.sort_values(by=[6031953227014493100], ascending=False)
```

```python
df_results.to_csv('ContentFiltering.csv')
```
